```python
import tensorflow as tf
import cv2
```

```python
img1 = cv2.imread('./images/dog0.jpg')
img2 = cv2.imread('./images/f0.png')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img1)
```

```python
import numpy as np
```

```python
plt.imshow(img2)
```

```python
model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(10, 3, activation='relu'),
    tf.keras.layers.GlobalAveragePooling2D(),
    tf.keras.layers.Dense(1)
])
```

```python
img1 = np.reshape(img1, (1,1440,1920,3))/255
```

```python
img2 = np.reshape(img2, (1,110,110,3))/255
```

```python
t = model(img1)
```

```python
model.summary()
```

```python
j = model(img2)
```

```python
print(img1.shape, img2.shape)
```

```python
print(t[0],j[0])
```
