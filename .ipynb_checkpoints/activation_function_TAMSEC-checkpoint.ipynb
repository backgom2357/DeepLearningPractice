```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from itertools import product
```

# TRAIN 

```python
X = [(0,0),(0,1),(1,0),(1,1)]
y = [ t[0] ^ t[1] for t in X]
```

```python
X = np.array(X).reshape(4,2)
y = np.array(y)
```

```python
px = X[:,0]
py = X[:,1]
area1 = np.ma.masked_where(y < 0.5, y)
area2 = np.ma.masked_where(y >= 0.5, y+1)
plt.figure(figsize=(7,7))
plt.scatter(px, py, s=area1*200, marker='^')
plt.scatter(px, py, s=area2*200, marker='o')
```

```python
model = tf.keras.Sequential([
    tf.keras.layers.Dense(2, 
                          activation='relu', 
                          input_shape = (2,)),
    tf.keras.layers.Dense(1)
    ])
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.01),
              loss='mse',
              metrics=['accuracy'])
```

```python
history = model.fit(X,y, epochs=100, verbose=0)
```

```python
model(X)
```

```python
pred = np.argmax(model(X), axis=1)
print(y)
print(pred)
print(y==pred)
```

# Let's ! Perceptron 

```python
import numpy as np

class Perceptron(object):

    def __init__(self, input_shape, threshold=100, learning_rate=0.01):
        self.learning_rate = learning_rate
        self.weights = np.zeros(input_shape)
        self.bias = np.zeros(1)
           
    def predict(self, inputs):
        summation = np.dot(inputs, self.weights[1:]) + self.weights[0]
        if summation > 0:
            activation = 1
        else:
            activation = 0            
        return activation

    def train(self, training_inputs, labels):
        for _ in range(self.threshold):
            for inputs, label in zip(training_inputs, labels):
                prediction = self.predict(inputs)
                self.weights[1:] += self.learning_rate * (label - prediction) * inputs
                self.weights[0] += self.learning_rate * (label - prediction)
```

```python
import numpy as np

training_inputs = []
training_inputs.append(np.array([1, 1]))
training_inputs.append(np.array([1, 0]))
training_inputs.append(np.array([0, 1]))
training_inputs.append(np.array([0, 0]))

labels = np.array([1, 0, 0, 0])

perceptron = Perceptron(2)
perceptron.train(training_inputs, labels)

inputs = np.array([1, 1])
print(perceptron.predict(inputs))
#=> 1

inputs = np.array([0, 1])
print(perceptron.predict(inputs)) 
#=> 0
```
